```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

--- Step 1: Load Data ---
 Reasoning: The first step is to load our dataset into a pandas DataFrame,
 which is the standard tool for data manipulation in Python.
 Make sure the CSV file from the dataset link is available.

```python
df = pd.read_csv(r"C:\Users\hp\Downloads\intern_data_ikarus.csv")
print("Dataset Head:")
print(df.head())
print("\nDataset Info:")
df.info()
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'].dropna(), kde=True, bins=50)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_brands = df['brand'].value_counts().nlargest(15)
sns.barplot(x=top_brands.values, y=top_brands.index, palette='viridis')
plt.title('Top 15 Brands in the Dataset')
plt.xlabel('Number of Products')
plt.ylabel('Brand')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_categories = df['categories'].value_counts().nlargest(15)
sns.barplot(x=top_categories.values, y=top_categories.index, palette='plasma')
plt.title('Top 15 Product Categories')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```

```python
analytics_data = {
    'top_brands': top_brands.to_dict(),
    'top_categories': top_categories.to_dict()
}
import json
with open(r'C:\Users\hp\Desktop\ikarus3d_project\backend\app\analytics_data.json', 'w') as f:
    json.dump(analytics_data, f)

print("\nAnalytics data saved for the backend.")
```
